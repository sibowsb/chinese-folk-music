# Fetching artist and song info using an external API

Reference:
[A pre-developed API for NetCloud Music](https://zhuanlan.zhihu.com/p/30246788) by `CeuiLiSA` on Zhihu, 2017.

```python
import urllib.request
import json
```

```python
api_base = 'https://api.imjad.cn/cloudmusic/'
data = {'type': 'artist', 'id': 2111}    # 崔健
request = api_base + '?' + '&'.join('{}={}'.format(k, data[k]) for k in data)
request
```

```python
response_raw = urllib.request.urlopen(request)
response_txt = response_raw.read().decode('utf-8')
response_json = json.loads(response_txt)
```
